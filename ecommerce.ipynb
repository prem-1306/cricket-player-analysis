```python
#RETAIL SALES DATA ANALYSIS
import pandas as pd


data = {
    "OrderID": list(range(1001, 1021)),
    "Customer": ["Amit", "Neha", "Ravi", "Priya", "Karan", "Suman", "Ankit", "Ravi", "Neha", "Amit",
                 "Priya", "Amit", "Neha", "Suman", "Karan", "Ravi", "Amit", "Priya", "Ankit", "Neha"],
    "Product": ["Laptop", "Phone", "Tablet", "Headphones", "Laptop", "Phone", "Smartwatch", "Laptop", "Tablet", "Phone",
                "Headphones", "Laptop", "Smartwatch", "Laptop", "Phone", "Tablet", "Phone", "Laptop", "Headphones", "Tablet"],
    "Category": ["Electronics", "Electronics", "Electronics", "Accessories", "Electronics",
                 "Electronics", "Accessories", "Electronics", "Electronics", "Electronics",
                 "Accessories", "Electronics", "Accessories", "Electronics", "Electronics",
                 "Electronics", "Electronics", "Electronics", "Accessories", "Electronics"],
    "Quantity": [1, 2, 1, 3, 1, 2, 1, 2, 2, 1,
                 3, 1, 1, 2, 1, 1, 2, 1, 4, 2],
    "Price": [60000, 20000, 15000, 2000, 62000, 21000, 8000, 58000, 15500, 22000,
              2500, 70000, 7500, 59000, 23000, 16000, 21000, 72000, 2000, 15000],
    "City": ["Delhi", "Mumbai", "Delhi", "Bangalore", "Chennai", "Delhi", "Mumbai", "Delhi", "Bangalore", "Chennai",
             "Mumbai", "Delhi", "Bangalore", "Chennai", "Delhi", "Mumbai", "Delhi", "Bangalore", "Chennai", "Delhi"],
    "Date": pd.date_range("2025-01-01", periods=20, freq="D")
}

df = pd.DataFrame(data)
df["Total_Amount"] = df["Quantity"] * df["Price"]
print(df)
print("\nTotal sales of the entire database:\n",df["Total_Amount"].sum())
print("\nTotal sales done in every city :\n",df.groupby("City")["Total_Amount"].sum())
print("\nAverage quantity of products purchased by each customer :\n",df.groupby("Customer")["Quantity"].mean().round(2))
top_3_cus=df.groupby("Customer")["Total_Amount"].sum()
print("\nTop 3 customer who spent the highest amount:\n",top_3_cus.nlargest(3))
popular_product_qty=df.groupby("Product")["Quantity"].sum()
print("\n The most popular product is :\n",popular_product_qty.idxmax(),"---",popular_product_qty.max())
print("\nTotal revenue generated bvy each product is :\n",df.groupby("Product")["Total_Amount"].sum())
high_rev_city=df.groupby("City")["Total_Amount"].sum()
print("\nThe city which have highest revenue is:\n",high_rev_city.idxmax(),"---",high_rev_city.max())
print("\n Average Order value of the city :\n",df.groupby("City")["Total_Amount"].mean().round(2))
max_order_no=df.groupby("Customer")["OrderID"].count()
print("\nThe customer who placed maximum number of order is :\n",max_order_no.idxmax(),"---",max_order_no.max())
high_avg_pro=df.groupby("Product")["Price"].mean()
print("\nThe product with highest average price is:\n",high_avg_pro.idxmax(),"---",high_avg_pro.max())
print("\nTotal Quantity sold for each product in each city:\n",df.groupby(["City","Product"])["Quantity"].sum())
print("\n Top 2 cities who generated highest revenue\n",high_rev_city.nlargest(2))
print("\n Average Total  amount spending of each customer in each city:\n",\
      df.groupby(["Customer","City"])["Total_Amount"].mean().round(2))
high_rev_pro=df.groupby(["City","Product"])["Total_Amount"].sum() 
print("\nRevenue of each product in each city:\n", high_rev_pro)
print("\nthe products who generated highest revenue in according to thier city is :")
for city in df["City"].unique():
    city_data=high_rev_pro[city]
    top_product=city_data.idxmax()
    top_revenue=city_data.max()
    print(f"{city} : {top_product}---{top_revenue}")
cust_pro_spending=df.groupby(["Customer","Product"])["Total_Amount"].sum()
print("\nCustomer spending total amount on every product:\n",cust_pro_spending)
max_spent_product = cust_pro_spending.groupby("Customer").idxmax()
print(max_spent_product)
max_spent_value = cust_pro_spending.groupby("Customer").max()
print(max_spent_value)
print("\n Customer wise most spent Product:")
for cust,prod in max_spent_product.items():
    print(f"\n {cust} : {prod[1]} ---- {max_spent_value[cust]}")
print("\n Average spending per order of each customer:\n")
print(df.groupby("Customer")["Total_Amount"].sum()/df.groupby("Customer")["OrderID"].count())
high_avg_rev_pro=df.groupby("Product")["Total_Amount"].mean()
print("\n Product who generated Highest Revenue :\n",high_avg_rev_pro.idxmax(),"---",high_avg_rev_pro.max())   

uni_pro=df.groupby("Customer")["Product"].nunique()
print("\nNo. of unique product bought by each customer:\n",uni_pro)
pro_count=df.groupby("Customer")["Quantity"].sum()
max_qty=pro_count.max()
top_cust=pro_count[pro_count==max_qty]
print("\nHighest no. of product brought by this customers:\n")
for cust,qty in top_cust.items():
    print(cust,"---",qty)
# print(pro_count)
# print("\nhighest no. of product bought by the customer is: \n",pro_count.idxmax(),"---",pro_count.max())
day_sales=df.groupby(df["Date"].dt.day)["Total_Amount"].sum()
print("Day Wise Sales:\n",day_sales)
print("Highest Sales Generated on Day:",day_sales.idxmax(),"---",day_sales.max()) 
week_sales=df.groupby(df["Date"].dt.isocalendar().week)["Total_Amount"].sum()
print("\nWeek Wise Sales:\n",week_sales)
print("Highest week sales happen on week :",week_sales.idxmax(),"---",week_sales.max())
print("\n Each customer spent average amount on each product:\n")
print(df.groupby(["Customer","Product"])["Total_Amount"].mean().round(2))
cat_high_rev=df.groupby("Category")["Total_Amount"].sum()
print("\n the Category which generated the highest reveue is :\n",cat_high_rev.idxmax(),"---",cat_high_rev.max())
print("\n The top 2 product who were sold at highest quantity are:\n",df.groupby("Product")["Quantity"].sum().nlargest(2))
high_rev_cust=df.groupby(["City","Customer"])["Total_Amount"].sum()
print("\ncity wise customer who generated highest revenue:\n")
for city in df["City"].unique():
    city_data=high_rev_cust[city]
    top_customer=city_data.idxmax()
    top_rev=city_data.max()
    print(f"{city}--{top_customer}--{top_rev}")

print("\nDESCRIPTIVE STATS OF DATA BASE:\n")
print(df[["OrderID","Quantity","Price","Total_Amount"]].describe().round(2))
df["Cumulative Sale"]=df["Total_Amount"].cumsum()
print(" Cumulative sale:\n",df["Cumulative Sale"])
cust_contri=(df.groupby("Customer")["Total_Amount"].sum()/df["Total_Amount"].sum()*100).round(2)
city_contri=(df.groupby("City")["Total_Amount"].sum()/df["Total_Amount"].sum()*100).round(2)
product_contri=(df.groupby("Product")["Total_Amount"].sum()/df["Total_Amount"].sum()*100).round(2)
category_contri=(df.groupby("Category")["Total_Amount"].sum()/df["Total_Amount"].sum()*100).round(2)
print("\nCustomer contribution for revenue:\n",cust_contri)
print("\nCity contribution for revenue:\n",city_contri)
print("\nProduct contribution for revenue:\n",product_contri)
print("\nCategory contribution for revenue:\n",category_contri)
print("\nMost Profitable product Category is :",category_contri.idxmax(),"---",category_contri.max())
cust_freq=df.groupby("Customer")["OrderID"].count()
max_freq=cust_freq.max()
loyal_cust=cust_freq[cust_freq==max_freq]
for cust,freq in loyal_cust.items():
    print(f"most loyal custometrs are: {cust}---{freq}")
#recency:har customer ne kitne din phele order kia tha
last_date=df["Date"].max()
recency=df.groupby("Customer")["Date"].max().apply(lambda x: (last_date-x).days)
#frequency
frequency=df.groupby("Customer")["OrderID"].count()
#monetory
monetory=df.groupby("Customer")["Total_Amount"].sum()
#RFM analysis
rfm=pd.DataFrame({"Recency(days)":recency,
                  "Frequency":frequency,
                  "Monetory":monetory})
print("Correlation Check:\n",df[["Quantity","Price","Total_Amount"]].corr())
rfm.to_csv("RFM.csv",index=False)
df.to_csv("retailsanalysis.csv",index=False)
```
